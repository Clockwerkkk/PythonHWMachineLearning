Suppose, we want to apply k-means for the following dataset with the five observations of two variables:
$$\{(0,11), (1,8), (0.1, 12), (1.1, 4), (1.2, 3)\}$$

Our random initialization ended up with the following cluster centers $C_1=(1.0,4.5)$ and $C_2=(0.6,11)$. Assume that, unlike ISLR, our algorithm starts with random cluster centroids instead of random cluster assignments (you can face such modification of k-means in UoL exams, also note that there multiple other ways of intializing and updating k-means clustering).

```python
df = dfRaw.query('IsTest==1').iloc[:,2:].fillna(0) # drop irrelevant features: IsTest and Y
df
```

```python
print(df.kurtosis()) #Kurtosis to see the distribution and detect outliers
```

```python
import seaborn as sns
```

```python
sns.displot(df["r(A_VI)(Å)"])
sns.displot(df["τ"])
sns.displot(df["l(A-O)(Å)"])
```

```python
q_low_ra, q_hi_ra = df["r(A_VI)(Å)"].quantile(0.02), df["r(A_VI)(Å)"].quantile(0.94) #Calculating quartiles
q_low_t, q_hi_t = df["τ"].quantile(0.035), df["τ"].quantile(0.98)
q_low_la, q_hi_la = df["l(A-O)(Å)"].quantile(0.02), df["l(A-O)(Å)"].quantile(0.98)
```

```python
dfnew, df_outliers_ra = df[(df["r(A_VI)(Å)"] < q_hi_ra) & (df["r(A_VI)(Å)"] > q_low_ra)], df[(df["r(A_VI)(Å)"] >= q_hi_ra) | (df["r(A_VI)(Å)"] <= q_low_ra)] #Split data with outliers into different dataframe
dfnew, df_outliers_t = dfnew[(dfnew["τ"] < q_hi_t) & (dfnew["τ"] > q_low_t)], dfnew[(dfnew["τ"] >= q_hi_t) | (dfnew["τ"] <= q_low_t)]
dfnew, df_outliers_la = dfnew[(dfnew["l(A-O)(Å)"] < q_hi_la) & (dfnew["l(A-O)(Å)"] > q_low_la)], dfnew[(df["l(A-O)(Å)"] >= q_hi_la) | (dfnew["l(A-O)(Å)"] <= q_low_la)]

outliers = pd.concat([df_outliers_ra,df_outliers_t,df_outliers_la])
```
